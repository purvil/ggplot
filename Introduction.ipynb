* Grammar of graphics
* First create a plot
```
ggplot(df, aes('xvar', 'yvar'))
```
* We have not said what kind of plot we want
* Let's add scatter plot
```
ggplot(df, aes('xvar', 'yvar')) + geom_point(color = 'red') # scatter plot
```

* Also add line plot
```
ggplot(df, aes('xvar', 'yvar')) + geom_point(color = 'red') + geom_line(color='red')
```

* add title and label

```
ggplot(df, aes('xvar', 'yvar')) + geom_point(color = 'red') + geom_line(color='red') + ggtitle('title') + xlab('x-label') + ylab('y-label')
```

````python
import pandas as pd
````

````python
from ggplot import *
````

````python
df = pd.read_csv('hr_year.csv')
````

````python
ggplot(df, aes('yearID', 'HR')) + geom_point(color='red')
````

````python
ggplot(df, aes('yearID', 'HR')) + geom_point(color='red') + geom_line(color='green') + \
ggtitle("Number of homerun hits over the year") + xlab("Years") + ylab("Total Homerun")
````
